```python
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# reading the csv file

#genero un url del documento de excel que voy a usar para el ejercicio , pues esta subido en mi github

url= "https://raw.githubusercontent.com/gespynoza/Visualizacion/main/population_total.csv"
df_population_raw = pd.read_csv(url)
```

# Making a Pivot Table

.pivot(): Returns reshaped DataFrame organized by given index / column values ("pivot without aggregation")

```python
# dropping null values
df_population_raw.dropna(inplace=True)
```

```python
#Count each country in the data set

counts = df_population_raw["country"].value_counts().sort_values(ascending=False)

# Print the result
print(counts)
```

```python
# making a pivot table
df_pivot = df_population_raw.pivot(index='year', columns='country',
                                        values='population')
```

```python
# showing pivot table
df_pivot
```

```python
# selecting some countries
cases = df_pivot[['United States', 'Costa Rica', 'China', 
                           'Germany', 'Portugal']]
```

```python
cases 
```

# Lineplot

```python
cases.plot(kind='line',
              xlabel='Years', ylabel='Population',
              title='Population by Country (1955-2020)',
              figsize=(8,4),legend=True )
```

```python
otrochart= cases.plot(kind="line", marker='.', linewidth=2,
              xlabel='Years', ylabel='Population',
              title='Population by Country (1955-2020)',
               figsize=(8,4),legend=True )


```

# Barplot

## Single Barplot

```python
# selecting only one year (2020)
df_pivot_2020 = cases[cases.index.isin([2020])]
```

```python
# showing new dataframe
df_pivot_2020
```

```python
# transpose dataframe from rows to columns or viceversa
df_trf_2020 = df_pivot_2020.T

```

```python
# making barplot
graphb=df_trf_2020.plot(kind='bar', color=(0.1, 0.5, 0.1, 0.1),  edgecolor='blue',
                   xlabel='Years', ylabel='Population',
                   title='Population in 2020', width = 0.8,fontsize = 12, alpha = 0.7)
```

```python
colors = sns.color_palette('pastel')
```

```python

plt.bar(cases['year'], cases['country'], color=colors[:5])
plt.xlabel('Years')
plt.ylabel('Population 2020')
plt.title('Barplot')
plt.show()
```
